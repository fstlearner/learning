```python
import pandas as pd
import numpy as np
```

```python
train=pd.read_csv('train.csv') #loading train dataset from the local machine.
test=pd.read_csv('test.csv')   #loading test dataset from the local machine
```

```python
train.head()
```

```python
test.head()
```

```python
missing_value=train.isnull().sum() #checking for missing values in dataset 
print(missing_value) #There is no any data is missing.
```

```python
columns_to_drop=['Hardness']
X=train.drop(columns=columns_to_drop) #Dropping the target and unaffected columns from the dataset.
X.head()
```

```python
Y=train['Hardness']
Y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.01)
```

```python
from sklearn.ensemble import RandomForestRegressor
model=RandomForestRegressor()
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_train,y_train)
```

```python
model.predict(test)
y_predict=model.predict(test)
y_predict
```

```python
df_test = pd.DataFrame(y_predict)
```

```python
df_test.to_csv('test_data.csv', index=False)
```
