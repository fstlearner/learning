```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
iris=load_iris()
dir(iris)
```

```python
iris = pd.DataFrame(iris.data,columns=iris.feature_names)
iris.head(60)
```

```python
inputs=iris.drop(['sepal length (cm)','sepal width (cm)'],axis='columns')
inputs
```

# Scalling Data


```python
#scalling data
from sklearn.preprocessing import MinMaxScaler #importing necessary library for dcalling.
scaler=MinMaxScaler()
scaler.fit(inputs[['petal length (cm)']])
inputs['petal length (cm)']=scaler.fit_transform(inputs['petal length (cm)'].values.reshape(-1, 1))
scaler.fit(inputs[['petal width (cm)']])
inputs['petal width (cm)']=scaler.fit_transform(inputs['petal width (cm)'].values.reshape(-1, 1))
inputs.head(60)
```

# Data visulization 

```python
import matplotlib.pyplot as plt
plt.scatter(iris['petal length (cm)'],iris['petal width (cm)']) #from plot we can see there is two cluster hence from the elbow method we can found that there is 3 cluster.
```

# Finding value of K


```python
sse=[]
k_rng=range(1,10)
for i in k_rng:
    km=KMeans(n_clusters=i,n_init=10)
    km.fit(inputs[['petal length (cm)','petal width (cm)']])
    sse.append(km.inertia_)
```

```python
sse
```

```python
plt.plot(k_rng,sse)
plt.xlabel('K_range')
plt.ylabel('SSE')
#from the graph it is clear that the data can be divided into 3 cluster.
```

# Training the model


```python
km = KMeans(n_clusters=3,n_init=10)
y_predict=km.fit_predict(inputs)
```

```python
inputs['clusters']=y_predict
```

# Data visualization

```python
df1=inputs[inputs.clusters==0]
df2=inputs[inputs.clusters==1]
df3=inputs[inputs.clusters==2]
```

```python
plt.scatter(df1['petal length (cm)'],df1['petal width (cm)'],color='blue')
plt.scatter(df2['petal length (cm)'],df2['petal width (cm)'],color='green')
plt.scatter(df3['petal length (cm)'],df3['petal width (cm)'],color='black')
```
